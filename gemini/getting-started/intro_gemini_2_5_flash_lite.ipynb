## 概述

Gemini 2.5 Flash-Lite 是 Google 迄今为止最具成本效益的 Gemini 2.5 模型，针对高吞吐量工作负载的性能进行了优化。它比之前的 2.0 Flash 和 Flash-Lite 模型具有更高的性能，并显著改善了延迟，非常适合分类、翻译、智能路由以及其他对成本敏感的大规模操作等任务。

### 目标

在本教程中，您将学习如何使用 Gemini API 和适用于 Python 的 Google Gen AI SDK 与 Gemini 2.5 Flash-Lite 模型。

您将完成以下任务：

- 生成文本
- 控制思考预算
- 查看摘要思考内容
- 设置模型参数和系统指令
- 使用安全过滤器
- 使用受控生成
- 将搜索作为工具使用
- 使用代码执行

Gemini 2.5 Flash-Lite 还支持其他标准 Gemini 功能，例如多模态输入、自动和手动函数调用、计算 token 数量、聊天补全等。这些功能不在本教程范围内。

### 导入库


````
from typing import List

from IPython.display import HTML, Markdown, display
from google import genai
from google.genai.types import (
    GenerateContentConfig,
    GoogleSearch,
    HarmBlockThreshold,
    HarmCategory,
    SafetySetting,
    ThinkingConfig,
    Tool,
    ToolCodeExecution,
)
from pydantic import BaseModel
````

### 连接到 Vertex AI 上的生成式 AI 服务

````
client = genai.Client(vertexai=True, project=PROJECT_ID, location=LOCATION)
````

## 使用 Gemini 2.5 Flash-Lite

了解有关 [Gemini 2.5 Flash-Lite 模型](https://cloud.google.com/vertex-ai/generative-ai/docs/models/gemini/2-5-flash-lite) 的更多信息。本笔记本中使用的模型 ID 为 `gemini-2.5-flash-lite`。

````
MODEL_ID = "gemini-2.5-flash-lite"  # @param {type: "string"}
````

### 发送您的第一个提示

- 使用 `generate_content` 方法生成对提示的响应。您可以将文本和其他多模态输入传递给 `generate_content`。
- 使用 `generate_content_stream` 方法在生成响应时对其进行流式传输，模型将在生成响应块后立即返回。
- 使用 `.text` 属性获取响应的文本内容。默认情况下，Gemini 使用 [Markdown](https://daringfireball.net/projects/markdown/) 语法输出格式化文本。

````
response = client.models.generate_content(
    model=MODEL_ID,
    contents="Why is the sky blue?",
)

display(Markdown(response.text))
````

### 设置模型参数和系统指令

- 您可以在每个模型请求中设置 [Gemini API 参数](https://cloud.google.com/vertex-ai/generative-ai/docs/model-reference/inference#parameters)，以控制模型生成响应的方式。
- [系统指令](https://cloud.google.com/vertex-ai/generative-ai/docs/learn/prompts/system-instruction-introduction) 为模型提供额外的上下文，以理解任务、提供更个性化的响应，并在用户交互过程中遵守相关准则。
- 您还可以配置 [安全过滤器](https://cloud.google.com/vertex-ai/generative-ai/docs/multimodal/configure-safety-filters)，以调整模型处理不同内容类别的方式。


````
system_instruction = """
Persona: You are 'The Curator,' a sophisticated AI historian. Your tone is eloquent and engaging.
Task: Create a thematic travel itinerary that creatively combines the user's interests.
Rules:
- Use Markdown for all formatting.
- Each day must have a ### Theme:.
- Every day must include these exact bolded sections: Morning:, Afternoon:, Evening:, The Curator's Note:, and Required Attire:.
- Directly connect the itinerary to all of the user's stated interests.
"""

prompt = """
Destination: Florence, Italy
Duration: 2 Days
Interests: Medici family banking and the use of light/optics in Renaissance art.
Goal: Create an itinerary showing how money and light converged to create Florentine masterpieces.
"""

safety_settings = [
    SafetySetting(
        category=HarmCategory.HARM_CATEGORY_DANGEROUS_CONTENT,
        threshold=HarmBlockThreshold.BLOCK_LOW_AND_ABOVE,
    ),
]

response = client.models.generate_content(
    model=MODEL_ID,
    contents=prompt,
    config=GenerateContentConfig(
        system_instruction=system_instruction,
        temperature=0.7,
        top_p=0.95,
        candidate_count=1,
        max_output_tokens=2048,
        safety_settings=safety_settings,
    ),
)

display(Markdown(response.text))
````

### 控制思考预算

Gemini 2.5 Flash-Lite 是一种混合推理模型，可以对任务应用扩展的思考过程，并使用工具来最大化响应准确性。与 Gemini 2.5 Flash 不同，Gemini 2.5 Flash-Lite 默认关闭思考模式。您可以设置可选的 `thinking_budget` 参数来控制模型针对给定提示的思考量。


- `thinking_budget=0`：关闭思考（默认）
- `thinking_budget=-1`：动态思考
- `thinking_budget=[512-24576]`：分配思考预算

然后使用 `generate_content` 或 `generate_content_stream` 方法发送请求，以使用 `ThinkingConfig` 生成内容。


````
THINKING_BUDGET = 1024  # @param {type: "integer"}

response = client.models.generate_content(
    model=MODEL_ID,
    contents="How many R's are in the word strawberry?",
    config=GenerateContentConfig(
        thinking_config=ThinkingConfig(
            include_thoughts=True,
            thinking_budget=THINKING_BUDGET,
        )
    ),
)

display(Markdown(response.text))
````

您也可以选择打印模型响应中的 `usage_metadata` 和 token 计数。

````
print(f"prompt_token_count: {response.usage_metadata.prompt_token_count}")
print(f"candidates_token_count: {response.usage_metadata.candidates_token_count}")
print(f"thoughts_token_count: {response.usage_metadata.thoughts_token_count}")
print(f"total_token_count: {response.usage_metadata.total_token_count}")
````

### 查看摘要思考内容

您可以将 `include_thoughts` 标志设置为 `True`，使模型在返回最终答案的同时也返回其"思考"摘要。当您使用此设置发出请求时，响应将包含多个部分。您可以检查 `part.thought` 字段，以识别哪个部分包含模型的思考内容。

````
response = client.models.generate_content(
    model=MODEL_ID,
    contents="How many R's are in the word strawberry?",
    config=GenerateContentConfig(
        thinking_config=ThinkingConfig(include_thoughts=True, thinking_budget=-1)
    ),
)

for part in response.candidates[0].content.parts:
    if part.thought:
        display(Markdown(f"**Thoughts**: {part.text}"))
    else:
        display(Markdown(f"**Answer**: {part.text}"))
````

## 使用受控生成

受控生成允许您定义响应架构，以指定模型输出的结构，包括字段名称和预期数据类型。模型的输出将严格遵循所提供的架构。

````
class Recipe(BaseModel):
    name: str
    description: str
    ingredients: list[str]


class RecipeList(BaseModel):
    recipes: List[Recipe]


response = client.models.generate_content(
    model=MODEL_ID,
    contents="List a few popular cookie recipes and their ingredients.",
    config=GenerateContentConfig(
        response_mime_type="application/json",
        response_schema=RecipeList,
    ),
)

recipes: List[Recipe] = response.parsed
for recipe in recipes.recipes:
    print(f"Recipe: {recipe.name}")
    print(f"Ingredients: {recipe.ingredients}")
    print("\n")
````

## 使用 Google 搜索进行信息基础增强（将搜索作为工具）

Google 搜索作为 Gemini 2.5 Flash-Lite 中的内置工具提供，这是 2.0 Flash-Lite 模型所不具备的功能。这使模型能够决定何时使用 Google 搜索来提高其响应的准确性和时效性。

````
google_search_tool = Tool(google_search=GoogleSearch())

response = client.models.generate_content(
    model=MODEL_ID,
    contents="What is the current temperature in Austin, TX?",
    config=GenerateContentConfig(
        tools=[google_search_tool],
    ),
)

display(Markdown(response.text))

HTML(response.candidates[0].grounding_metadata.search_entry_point.rendered_content)
````

## 代码执行

Gemini API 的代码执行功能使模型能够生成并运行 Python 代码。模型可以从代码运行结果中迭代学习，从而得出最终输出。

````
code_execution_tool = Tool(code_execution=ToolCodeExecution())

response = client.models.generate_content(
    model=MODEL_ID,
    contents="Calculate 20th fibonacci number. Then find the nearest palindrome to it.",
    config=GenerateContentConfig(
        tools=[code_execution_tool],
        temperature=0,
    ),
)

display(
    Markdown(
        f"""
## Code
```py
{response.executable_code}
```
### Output
```
{response.code_execution_result}
```
"""
    )
)
````
