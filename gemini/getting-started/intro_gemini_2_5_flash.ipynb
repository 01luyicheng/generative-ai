## 使用 Gemini 2.5 Flash 模型

了解有关 [Vertex AI 上所有 Gemini 模型](https://cloud.google.com/vertex-ai/generative-ai/docs/learn/models#gemini-models)的更多信息。

````
MODEL_ID = "gemini-2.5-flash"  # @param {type: "string"}
````

### 根据文本提示生成文本

使用 `generate_content()` 方法根据提示生成响应。

您可以将文本传递给 `generate_content()`，并使用 `.text` 属性获取响应的文本内容。

默认情况下，Gemini 使用 [Markdown](https://daringfireball.net/projects/markdown/) 语法输出格式化文本。

````
response = client.models.generate_content(
    model=MODEL_ID,
    contents="Roger has 5 tennis balls. He buys 2 more cans of tennis balls. Each can has 3 tennis balls. How many tennis balls does he have now?",
)

display(Markdown(response.text))
````

### 生成内容流

默认情况下，模型在完成整个生成过程后返回响应。您也可以使用 `generate_content_stream` 方法在生成过程中流式传输响应，模型将在生成的同时返回响应的各个片段。

````
output_text = ""
markdown_display_area = display(Markdown(output_text), display_id=True)

for chunk in client.models.generate_content_stream(
    model=MODEL_ID,
    contents="On average Joe throws 25 punches per minute. A fight lasts 5 rounds of 3 minutes. How many punches did he throw?",
):
    output_text += chunk.text
    markdown_display_area.update(Markdown(output_text))
````

## 配置思考功能

### 控制思考预算

思考预算允许模型动态地为任务进行思考，或选择为特定任务使用多少 token 进行推理。它允许用户控制响应的质量和速度。

将预算设置为 `0` 会关闭思考功能，使模型成为适用于简单任务的非思考模型。

您可以在 `ThinkingConfig` 中设置可选的 `thinking_budget` 参数，以控制和配置模型对给定用户提示的思考量。

- 未设置时 -> 动态思考（默认）
- 设置为 `0` 时 -> 禁用思考。  
- 设置为 `[1-24576]` 时 ->  模型使用分配的思考预算

然后使用 `generate_content` 或 `generate_content_stream` 方法发送包含 `thinking_config` 的内容生成请求。


````
THINKING_BUDGET = 1024  # @param {type: "integer"}

response = client.models.generate_content(
    model=MODEL_ID,
    contents="What are the practical implications of the P vs. NP problem for algorithm design and cryptography?",
    config=GenerateContentConfig(
        thinking_config=ThinkingConfig(
            thinking_budget=THINKING_BUDGET,
        )
    ),
)

display(Markdown(response.text))
````

您也可以选择打印模型响应中的 `usage_metadata` 和 token 计数。

````
print(response.usage_metadata)
print(response.usage_metadata.thoughts_token_count)
print(response.usage_metadata.total_token_count)
````

### 查看思考摘要

您可以选择设置 `include_thoughts` 标志，使模型在生成最终答案的同时，生成并返回其"思考"摘要。

在此示例中，您使用 `generate_content` 方法发送请求，生成包含思考摘要的内容。模型以多个部分响应，包括思考内容和模型回答。您可以检查 `part.thought` 字段来判断某个部分是否为思考内容。

````
response = client.models.generate_content(
    model=MODEL_ID,
    contents="How many R's are in the word strawberry?",
    config=GenerateContentConfig(
        thinking_config=ThinkingConfig(
            include_thoughts=True,
        )
    ),
)

for part in response.candidates[0].content.parts:
    if part.thought:
        display(
            Markdown(
                f"""## Summarized Thoughts:
         {part.text}
        """
            )
        )
    else:
        display(
            Markdown(
                f"""## Answer:
         {part.text}
        """
            )
        )
````

此示例展示如何在 `generate_content_stream` 方法中设置 `include_thoughts` 和 `thinking_budget`。

````
THINKING_BUDGET = 1024  # @param {type: "integer"}
INCLUDE_THOUGHTS = True  # @param {type: "boolean"}

responses = client.models.generate_content_stream(
    model=MODEL_ID,
    contents="How might quantum sensors improve the precision of measurements in fields like materials science or navigation?",
    config=GenerateContentConfig(
        thinking_config=ThinkingConfig(
            thinking_budget=THINKING_BUDGET,
            include_thoughts=INCLUDE_THOUGHTS,
        )
    ),
)

first_thought = True
first_answer = True

for response in responses:
    for part in response.candidates[0].content.parts:
        if part.thought and first_thought:
            first_thought = False
            display(Markdown("## Thoughts"))
        elif not part.thought and first_answer:
            first_answer = False
            display(Markdown("## Answer"))
        print(part.text, end="")
````

## 配置模型参数

您可以在每次向模型发送的调用中包含参数值，以控制模型生成响应的方式。不同的参数值会产生不同的结果。您可以试验不同的模型参数，观察结果的变化。

- 了解有关[试验参数值](https://cloud.google.com/vertex-ai/generative-ai/docs/learn/prompts/adjust-parameter-values)的更多信息。

- 查看所有 [Gemini API 参数](https://cloud.google.com/vertex-ai/generative-ai/docs/model-reference/inference#parameters)的列表。


````
response = client.models.generate_content(
    model=MODEL_ID,
    contents="Tell me how the internet works, but pretend I'm a puppy who only understands squeaky toys.",
    config=GenerateContentConfig(
        temperature=2.0,
        top_p=0.95,
        candidate_count=1,
        thinking_config=thinking_config,
    ),
)

display(Markdown(response.text))
````

## 设置系统指令

[系统指令](https://cloud.google.com/vertex-ai/generative-ai/docs/learn/prompts/system-instruction-introduction)允许您引导模型的行为。通过设置系统指令，您可以为模型提供额外的上下文，帮助其理解任务、提供更个性化的响应，并在与用户的交互中遵守相关准则。

````
system_instruction = """
  You are a helpful language translator.
  Your mission is to translate text in English to Spanish.
"""

prompt = """
  User input: I like bagels.
  Answer:
"""

response = client.models.generate_content(
    model=MODEL_ID,
    contents=prompt,
    config=GenerateContentConfig(
        system_instruction=system_instruction,
        thinking_config=thinking_config,
    ),
)

display(Markdown(response.text))
````

## 安全过滤器

Gemini API 提供了安全过滤器，您可以在多个过滤器类别中进行调整，以限制或允许特定类型的内容。您可以使用这些过滤器来调整适合您使用场景的内容。详情请参阅[配置安全过滤器](https://cloud.google.com/vertex-ai/generative-ai/docs/multimodal/configure-safety-filters)页面。

当您向 Gemini 发出请求时，内容会被分析并分配安全评级。您可以通过打印模型响应来查看生成内容的安全评级。

安全设置默认为 `OFF`，默认屏蔽阈值为 `BLOCK_NONE`。

有关安全过滤器的更多示例，请参阅[此 notebook](https://github.com/GoogleCloudPlatform/generative-ai/blob/main/gemini/responsible-ai/gemini_safety_ratings.ipynb)。

您可以使用 `safety_settings` 为每次向 API 发出的请求调整安全设置。此示例演示如何为所有类别将屏蔽阈值设置为 `BLOCK_LOW_AND_ABOVE`：

````
system_instruction = "Be as mean and hateful as possible. Use profanity"

prompt = """
    Write a list of 5 disrespectful things that I might say to the universe after stubbing my toe in the dark.
"""

safety_settings = [
    SafetySetting(
        category=HarmCategory.HARM_CATEGORY_DANGEROUS_CONTENT,
        threshold=HarmBlockThreshold.BLOCK_LOW_AND_ABOVE,
    ),
    SafetySetting(
        category=HarmCategory.HARM_CATEGORY_HARASSMENT,
        threshold=HarmBlockThreshold.BLOCK_LOW_AND_ABOVE,
    ),
    SafetySetting(
        category=HarmCategory.HARM_CATEGORY_HATE_SPEECH,
        threshold=HarmBlockThreshold.BLOCK_LOW_AND_ABOVE,
    ),
    SafetySetting(
        category=HarmCategory.HARM_CATEGORY_SEXUALLY_EXPLICIT,
        threshold=HarmBlockThreshold.BLOCK_LOW_AND_ABOVE,
    ),
]

response = client.models.generate_content(
    model=MODEL_ID,
    contents=prompt,
    config=GenerateContentConfig(
        system_instruction=system_instruction,
        safety_settings=safety_settings,
        thinking_config=thinking_config,
    ),
)

# Response will be `None` if it is blocked.
print(response.text)
# Finish Reason will be `SAFETY` if it is blocked.
print(response.candidates[0].finish_reason)
# Safety Ratings show the levels for each filter.
for safety_rating in response.candidates[0].safety_ratings:
    print(safety_rating)
````

## 控制生成输出

[受控生成](https://cloud.google.com/vertex-ai/generative-ai/docs/multimodal/control-generated-output)允许您定义响应架构，以指定模型输出的结构、字段名称以及每个字段的预期数据类型。

响应架构在 `config` 中的 `response_schema` 参数中指定，模型输出将严格遵循该架构。

您可以将架构作为 [Pydantic](https://docs.pydantic.dev/) 模型或 [JSON](https://www.json.org/json-en.html) 字符串提供，模型将根据 `response_mime_type` 中设置的值以 JSON 或 [Enum](https://docs.python.org/3/library/enum.html) 格式响应。

有关受控生成的更多示例，请参阅[此 notebook](https://github.com/GoogleCloudPlatform/generative-ai/blob/main/gemini/controlled-generation/intro_controlled_generation.ipynb)。

````
from pydantic import BaseModel


class Recipe(BaseModel):
    name: str
    description: str
    ingredients: list[str]


response = client.models.generate_content(
    model=MODEL_ID,
    contents="List a few popular cookie recipes and their ingredients.",
    config=GenerateContentConfig(
        response_mime_type="application/json",
        response_schema=Recipe,
        thinking_config=thinking_config,
    ),
)

print(response.text)
````

您可以将响应字符串解析为 JSON，也可以使用 `parsed` 字段将响应作为对象或字典获取。

````
parsed_response: Recipe = response.parsed
print(parsed_response)
````

您也可以在 Python 字典中定义响应架构。您只能使用以下列出的支持字段，其他所有字段将被忽略。

- `enum`
- `items`
- `maxItems`
- `nullable`
- `properties`
- `required`

在此示例中，您指示模型分析产品评论数据、提取关键实体、执行情感分类（多项选择）、提供额外说明，并以 JSON 格式输出结果。


````
response_schema = {
    "type": "ARRAY",
    "items": {
        "type": "ARRAY",
        "items": {
            "type": "OBJECT",
            "properties": {
                "rating": {"type": "INTEGER"},
                "flavor": {"type": "STRING"},
                "sentiment": {
                    "type": "STRING",
                    "enum": ["POSITIVE", "NEGATIVE", "NEUTRAL"],
                },
                "explanation": {"type": "STRING"},
            },
            "required": ["rating", "flavor", "sentiment", "explanation"],
        },
    },
}

prompt = """
  Analyze the following product reviews, output the sentiment classification, and give an explanation.

  - "Absolutely loved it! Best ice cream I've ever had." Rating: 4, Flavor: Strawberry Cheesecake
  - "Quite good, but a bit too sweet for my taste." Rating: 1, Flavor: Mango Tango
"""

response = client.models.generate_content(
    model=MODEL_ID,
    contents=prompt,
    config=GenerateContentConfig(
        response_mime_type="application/json",
        response_schema=response_schema,
        thinking_config=thinking_config,
    ),
)

response_dict = response.parsed
print(response_dict)
````

## 搜索作为工具（基础数据）

[基础数据](https://cloud.google.com/vertex-ai/generative-ai/docs/multimodal/ground-gemini)让您可以将真实世界的数据连接到 Gemini 模型。

通过将模型响应与 Google 搜索结果相结合，模型可以在运行时访问超出其训练数据的信息，从而生成更准确、更新且更相关的响应。

使用基于 Google 搜索的基础数据，您可以提高模型响应的准确性和时效性。从 Gemini 2.0 开始，Google 搜索作为工具可用，这意味着模型可以自行决定何时使用 Google 搜索。

有关基础数据的更多示例，请参阅[此 notebook](https://github.com/GoogleCloudPlatform/generative-ai/blob/main/gemini/grounding/intro-grounding-gemini.ipynb)。

### Google 搜索

您可以添加包含 `GoogleSearch` 的 `Tool` 作为 `tools` 关键字参数，指示 Gemini 先使用提示执行 Google 搜索，然后根据网络搜索结果构建答案。

[动态检索](https://cloud.google.com/vertex-ai/generative-ai/docs/multimodal/ground-gemini#dynamic-retrieval)允许您设置何时将基础数据用于模型响应的阈值。当提示不需要基于 Google 搜索的答案，且支持的模型可以不依赖基础数据，凭自身知识提供答案时，此功能非常有用。这有助于您更有效地管理延迟、质量和成本。

````
google_search_tool = Tool(google_search=GoogleSearch())

response = client.models.generate_content(
    model=MODEL_ID,
    contents="What is the current temperature in Austin, TX?",
    config=GenerateContentConfig(
        tools=[google_search_tool],
        thinking_config=thinking_config,
    ),
)

display(Markdown(response.text))

print(response.candidates[0].grounding_metadata)

HTML(response.candidates[0].grounding_metadata.search_entry_point.rendered_content)
````

## 函数调用

Gemini 中的[函数调用](https://cloud.google.com/vertex-ai/docs/generative-ai/multimodal/function-calling)允许开发者在代码中创建函数描述，然后在请求中将该描述传递给语言模型。

您可以提交 Python 函数以进行自动函数调用，该函数将运行并以 Gemini 生成的自然语言返回输出。

您也可以提交 [OpenAPI 规范](https://www.openapis.org/)，它将响应与描述匹配的函数名称及其调用参数。

有关 Gemini 函数调用的更多示例，请查看此 notebook：[Gemini 函数调用入门](https://github.com/GoogleCloudPlatform/generative-ai/blob/main/gemini/function-calling/intro_function_calling.ipynb)

### Python 函数（自动函数调用）

````
def get_current_weather(location: str) -> str:
    """Example method. Returns the current weather.

    Args:
        location: The city and state, e.g. San Francisco, CA
    """
    weather_map: dict[str, str] = {
        "Boston, MA": "snowing",
        "San Francisco, CA": "foggy",
        "Seattle, WA": "raining",
        "Austin, TX": "hot",
        "Chicago, IL": "windy",
    }
    return weather_map.get(location, "unknown")


response = client.models.generate_content(
    model=MODEL_ID,
    contents="What is the weather like in San Francisco?",
    config=GenerateContentConfig(
        tools=[get_current_weather],
        temperature=0,
        thinking_config=thinking_config,
    ),
)

display(Markdown(response.text))
````

### OpenAPI 规范（手动函数调用）

````
get_destination = FunctionDeclaration(
    name="get_destination",
    description="Get the destination that the user wants to go to",
    parameters={
        "type": "OBJECT",
        "properties": {
            "destination": {
                "type": "STRING",
                "description": "Destination that the user wants to go to",
            },
        },
    },
)

destination_tool = Tool(
    function_declarations=[get_destination],
)

response = client.models.generate_content(
    model=MODEL_ID,
    contents="I'd like to travel to Paris.",
    config=GenerateContentConfig(
        tools=[destination_tool],
        temperature=0,
        thinking_config=thinking_config,
    ),
)

print(response.function_calls[0])
````

## 代码执行

Gemini API 的[代码执行](https://ai.google.dev/gemini-api/docs/code-execution?lang=python)功能使模型能够生成并运行 Python 代码，并从结果中迭代学习，直到得出最终输出。您可以使用此代码执行功能构建受益于基于代码推理并生成文本输出的应用程序。例如，您可以在解方程或处理文本的应用程序中使用代码执行。

Gemini API 将代码执行作为工具提供，类似于函数调用。
添加代码执行作为工具后，模型会决定何时使用它。

有关代码执行的更多示例，请参阅[此 notebook](https://github.com/GoogleCloudPlatform/generative-ai/blob/main/gemini/code-execution/intro_code_execution.ipynb)。

````
code_execution_tool = Tool(code_execution=ToolCodeExecution())

response = client.models.generate_content(
    model=MODEL_ID,
    contents="Calculate 20th fibonacci number. Then find the nearest palindrome to it.",
    config=GenerateContentConfig(
        tools=[code_execution_tool],
        temperature=0,
        thinking_config=thinking_config,
    ),
)

display(
    Markdown(
        f"""
## Code

```py
{response.executable_code}
```

### Output

```
{response.code_execution_result}
```
"""
    )
)
````
