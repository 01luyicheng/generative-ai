## 概述

**Gemini 3.1 Pro** 是 Google 最新的旗舰模型。主要架构和性能改进包括：

* **增强的稳定性与事实依据**：提升事实准确性，减少重复的响应模式   
* **先进的编程与智能体能力：** 在软件工程和智能体性能方面取得了显著提升。  
* **效率与推理模式：** 提升推理任务的 token 效率，引入中等思考级别以平衡深度与延迟。  
* **核心质量改进：** 在通用推理、指令遵循和创意写作质量方面持续进步。

本笔记本是开发者通过 Vertex AI 上的 Google Gen AI SDK 与 **Gemini 3.1 Pro** 模型进行交互的快速入门指南。旨在演示模型的核心能力并展示新的 API 功能。

## 🚨 新 API 功能

### 1️⃣ 思考级别

`thinking_level` 参数允许您为模型的响应生成指定思考预算。通过选择不同的状态，您可以明确地在响应质量/推理复杂度与延迟/成本之间进行权衡。

- **低（Low）**：最小化延迟和成本。最适合简单的指令遵循或聊天。对应 `1` 到 `1,000` 之间的 token 预算。
- **中（Medium）**：在推理深度与延迟之间取得平衡。对应 `1,001` 到 `16,384` 之间的 token 预算。
- **高（High）**（默认）：最大化推理深度。模型可能需要更长时间才能生成第一个 token，但输出经过更充分的验证。对应 `16,384` 到 `32,768` 之间的 token 预算。

#### ⚠️ 注意事项

- 如果未指定 `thinking_level`，模型默认使用 `high`，根据提示词复杂度动态分配 token 预算。
- 将 token 预算设置为 `0` 将导致错误。
- 为防止循环，即使思考块为空，模型也会在多步骤交互的每个步骤中输出一个思考块。
- 不能在同一请求中同时使用 `thinking_level` 和旧版 `thinking_budget` 参数，否则将返回 400 错误。
- OpenAI Chat Completions API 的 `reasoning_effort` `medium` 对应 `thinking_level` 的 `high`。
- Gemini 3.1 Pro 无法关闭思考功能。


````
prompt = """
You are tasked with implementing the classic Thread-Safe Double-Checked Locking (DCL) Singleton pattern in modern C++. This task is non-trivial and requires specialized concurrency knowledge to prevent memory reordering issues.

Write a complete, runnable C++ program named `dcl_singleton.cpp` that defines a class `Singleton` with a private constructor and a static `getInstance()` method.

Your solution MUST adhere to the following strict constraints:
1. The Singleton instance pointer (`static Singleton*`) must be wrapped in `std::atomic` to correctly manage memory visibility across threads.
2. The `getInstance()` method must use `std::memory_order_acquire` when reading the instance pointer in the outer check.
3. The instance creation and write-back must use `std::memory_order_release` when writing to the atomic pointer.
4. A standard `std::mutex` must be used only to protect the critical section (the actual instantiation).
5. The `main` function must demonstrate safe, concurrent access by launching at least three threads, each calling `Singleton::getInstance()`, and printing the address of the returned instance to prove all threads received the same object.
"""

response = client.models.generate_content(
    model=MODEL_ID,
    contents=prompt,
    config=types.GenerateContentConfig(
        thinking_config=types.ThinkingConfig(
            thinking_level=types.ThinkingLevel.HIGH  # Dynamic thinking for high reasoning tasks
        )
    ),
)

display(Markdown(response.text))
````

### 2️⃣ 媒体分辨率

Gemini 3 通过 `media_resolution` 参数对多模态视觉处理引入了精细控制。更高的分辨率能提升模型读取细小文字或识别细节的能力，但会增加 token 用量和延迟。`media_resolution` 参数决定了每张输入图像或视频帧分配的最大 token 数量。

**⚠️ 注意**：由于媒体分辨率直接影响 token 数量，对于非常长的输入（如长视频或大量文档），您可能需要降低分辨率（例如设置为 `low`）以适应输入长度。

您可以为每个单独的媒体部分设置分辨率为 `low`、`medium` 或 `high`，示例如下：

````
response = client.models.generate_content(
    model=MODEL_ID,
    contents=[
        types.Part(
            file_data=types.FileData(
                file_uri="gs://cloud-samples-data/generative-ai/image/a-man-and-a-dog.png",
                mime_type="image/jpeg",
            ),
            media_resolution=types.PartMediaResolution(
                level=types.PartMediaResolutionLevel.MEDIA_RESOLUTION_HIGH
            ),
        ),
        types.Part(
            file_data=types.FileData(
                file_uri="gs://cloud-samples-data/generative-ai/video/behind_the_scenes_pixel.mp4",
                mime_type="video/mp4",
            ),
            media_resolution=types.PartMediaResolution(
                level=types.PartMediaResolutionLevel.MEDIA_RESOLUTION_LOW
            ),
        ),
        "When does the image appear in the video? What is the context?",
    ],
)

display(Markdown(response.text))
````

或者通过 `GenerateContentConfig` 全局设置。如果未指定，模型将根据媒体类型使用最优默认值。

````
response = client.models.generate_content(
    model=MODEL_ID,
    contents=[
        types.Part(
            file_data=types.FileData(
                file_uri="gs://cloud-samples-data/generative-ai/image/a-man-and-a-dog.png",
                mime_type="image/jpeg",
            ),
        ),
        "What is in the image?",
    ],
    config=types.GenerateContentConfig(
        media_resolution=types.MediaResolution.MEDIA_RESOLUTION_LOW
    ),
)

display(Markdown(response.text))
````

### 3️⃣ 思考签名

[思考签名](https://docs.cloud.google.com/vertex-ai/generative-ai/docs/thinking#signatures) 是加密 token，用于在多轮对话中（特别是使用函数调用时）保留模型的推理状态。

当思考模型决定调用外部工具时，它会暂停内部推理过程。思考签名充当“保存状态”的作用，使模型在您提供函数结果后能够无缝恢复其推理链。

Gemini 3 对最初在 Gemini 2.5 中引入的思考签名强制执行更严格的验证和更新的处理方式。为确保模型在多轮对话中保持上下文，您必须在后续请求中返回思考签名。


#### 自动处理思考签名（推荐）

如果您使用 Google Gen AI SDK（Python、Node.js、Go、Java）或 OpenAI Chat Completions API，并使用标准聊天历史功能或附加完整的模型响应，思考签名将自动处理。您无需对代码进行任何修改。


#### **示例 1**：自动函数调用

在 Gen AI SDK 中使用自动函数调用时，思考签名会自动处理。


````
def get_weather(city: str) -> str:
    """Gets the weather in a city."""
    if "london" in city.lower():
        return "Rainy"
    if "new york" in city.lower():
        return "Sunny"
    return "Cloudy"


response = client.models.generate_content(
    model=MODEL_ID,
    contents="What's the weather in London and New York?",
    config=types.GenerateContentConfig(
        tools=[get_weather],
    ),
)

# The SDK handles the function calls and thought signatures, and returns the final text
print("Final response:", response.text)

# Print function calling history
hist_turn = response.automatic_function_calling_history[1]
print("\nFunction Call 1:", hist_turn.parts[1].function_call.name)
````

#### **示例 2**：手动函数调用

在 Gen AI SDK 中使用手动函数调用时，如果您在顺序模型请求中附加完整的模型响应，思考签名也会自动处理。

````
# 1. Define your tool
get_weather_declaration = types.FunctionDeclaration(
    name="get_weather",
    description="Gets the current weather temperature for a given location.",
    parameters={
        "type": "object",
        "properties": {"location": {"type": "string"}},
        "required": ["location"],
    },
)
get_weather_tool = types.Tool(function_declarations=[get_weather_declaration])

# 2. Send a message that triggers the tool
prompt = "What's the weather like in London?"
response = client.models.generate_content(
    model=MODEL_ID,
    contents=prompt,
    config=types.GenerateContentConfig(
        tools=[get_weather_tool],
        thinking_config=types.ThinkingConfig(include_thoughts=True),
    ),
)

# 3. Handle the function call
function_call = response.function_calls[0]
location = function_call.args["location"]
print(f"Model wants to call: {function_call.name}")

# Execute your tool (e.g., call an API)
# (This is a mock response for the example)
print(f"Calling external tool for: {location}")
function_response_data = {
    "location": location,
    "temperature": "30C",
}

# 4. Send the tool's result back
# Append this turn's messages to history for a final response.
# The `content` object automatically attaches the required thought_signature behind the scenes.
history = [
    types.Content(role="user", parts=[types.Part(text=prompt)]),
    response.candidates[0].content,  # Signature preserved here
    types.Content(
        role="tool",
        parts=[
            types.Part.from_function_response(
                name=function_call.name,
                response=function_response_data,
            )
        ],
    ),
]

response_2 = client.models.generate_content(
    model=MODEL_ID,
    contents=history,
    config=types.GenerateContentConfig(
        tools=[get_weather_tool],
        thinking_config=types.ThinkingConfig(include_thoughts=True),
    ),
)

# 5. Get the final, natural-language answer
print(f"\nFinal model response: {response_2.text}")
````

#### 手动处理思考签名

如果您直接与 API 交互或管理原始 JSON 负载，必须正确处理模型响应中包含的 `thought_signature`。
在将对话历史发送回时，必须在接收到该签名的确切部分中返回此签名。

⚠️ 如果未返回正确的签名，模型将返回 400 错误 `"Function Call in the content block is missing a thought_signature"`。

更多详情请参阅[文档](https://docs.cloud.google.com/vertex-ai/generative-ai/docs/multimodal/function-calling#thinking)。

### 4️⃣ 流式函数调用

您可以使用流式部分函数调用参数来改善工具使用时的流式体验。通过将 `stream_function_call_arguments` 显式设置为 true 可启用此功能。

````
get_weather_declaration = types.FunctionDeclaration(
    name="get_weather",
    description="Gets the current weather temperature for a given location.",
    parameters={
        "type": "object",
        "properties": {"location": {"type": "string"}},
        "required": ["location"],
    },
)
get_weather_tool = types.Tool(function_declarations=[get_weather_declaration])


for chunk in client.models.generate_content_stream(
    model=MODEL_ID,
    contents="What's the weather in London and New York?",
    config=types.GenerateContentConfig(
        tools=[get_weather_tool],
        tool_config=types.ToolConfig(
            function_calling_config=types.FunctionCallingConfig(
                mode=types.FunctionCallingConfigMode.AUTO,
                stream_function_call_arguments=True,
            )
        ),
    ),
):
    function_call = chunk.function_calls[0]
    if function_call and function_call.name:
        print(f"{function_call.name}")
        print(f"will_continue={function_call.will_continue}")
````

### 5️⃣ 多模态函数响应

多模态函数调用允许用户在函数响应中包含多模态对象，从而改善对模型函数调用能力的利用。目前函数调用仅支持基于文本的函数响应。


````
# 1. Define the function tool
get_image_declaration = types.FunctionDeclaration(
    name="get_image",
    description="Retrieves the image file reference for a specific order item.",
    parameters={
        "type": "object",
        "properties": {
            "item_name": {
                "type": "string",
                "description": "The name or description of the item ordered (e.g., 'green shirt').",
            }
        },
        "required": ["item_name"],
    },
)
tool_config = types.Tool(function_declarations=[get_image_declaration])

# 2. Send a message that triggers the tool
prompt = "Show me the green shirt I ordered last month."
response_1 = client.models.generate_content(
    model=MODEL_ID,
    contents=[prompt],
    config=types.GenerateContentConfig(
        tools=[tool_config],
    ),
)

# 3. Handle the function call
function_call = response_1.function_calls[0]
requested_item = function_call.args["item_name"]
print(f"Model wants to call: {function_call.name}")

# Execute your tool (e.g., call an API)
# (This is a mock response for the example)
print(f"Calling external tool for: {requested_item}")

function_response_data = {
    "image_ref": {"$ref": "dress.jpg"},
}

function_response_multimodal_data = types.FunctionResponsePart(
    file_data=types.FunctionResponseFileData(
        mime_type="image/png",
        display_name="dress.jpg",
        file_uri="gs://cloud-samples-data/generative-ai/image/dress.jpg",
    )
)

# 4. Send the tool's result back
# Append this turn's messages to history for a final response.
history = [
    types.Content(role="user", parts=[types.Part(text=prompt)]),
    response_1.candidates[0].content,
    types.Content(
        role="tool",
        parts=[
            types.Part.from_function_response(
                name=function_call.name,
                response=function_response_data,
                parts=[function_response_multimodal_data],
            )
        ],
    ),
]

response_2 = client.models.generate_content(
    model=MODEL_ID,
    contents=history,
    config=types.GenerateContentConfig(
        tools=[tool_config],
        thinking_config=types.ThinkingConfig(include_thoughts=True),
    ),
)

print(f"\nFinal model response: {response_2.text}")
````

虽然以前的模型通常受益于调整 `temperature` 来控制创造性与确定性，但 Gemini 3 的推理能力已针对默认设置进行了优化。

````
prompt = """
A bat and a ball cost $1.10 in total. The bat costs $1.00 more than the ball.
How much does the ball cost?
"""

for chunk in client.models.generate_content_stream(
    model=MODEL_ID,
    contents=prompt,
    config=types.GenerateContentConfig(
        thinking_config=types.ThinkingConfig(
            thinking_level=types.ThinkingLevel.LOW,
        )
    ),
):
    print(chunk.text, end="")
````

### ✅ 思考摘要

您可以通过将 `include_thoughts` 设置为 `true` 在模型响应中包含思考摘要。


````
response = client.models.generate_content(
    model=MODEL_ID,
    contents="How many R's are in the word strawberry?",
    config=types.GenerateContentConfig(
        thinking_config=types.ThinkingConfig(
            include_thoughts=True,
            thinking_level=types.ThinkingLevel.LOW,
        )
    ),
)

for part in response.candidates[0].content.parts:
    if part.thought:
        display(
            Markdown(
                f"""## Thoughts:
         {part.text}
        """
            )
        )
    else:
        display(
            Markdown(
                f"""## Answer:
         {part.text}
        """
            )
        )
````

使用此辅助函数打印完整引用。

````
def print_grounding_data(response: types.GenerateContentResponse) -> None:
    """Print Gemini response with grounding citations in Markdown format."""
    candidate = response.candidates[0] if response.candidates else None
    metadata = getattr(candidate, "grounding_metadata", None)

    if not metadata:
        print("Response does not contain grounding metadata.")
        display(Markdown(response.text))
        return

    ENCODING = "utf-8"
    text_bytes = response.text.encode(ENCODING)
    parts = []
    last = 0

    # Insert citation markers into text
    for support in metadata.grounding_supports or []:
        end = support.segment.end_index
        parts.append(text_bytes[last:end].decode(ENCODING))
        parts.append(
            " " + "".join(f"[{i + 1}]" for i in support.grounding_chunk_indices)
        )
        last = end

    parts.append(text_bytes[last:].decode(ENCODING))
    parts.append("\n\n----\n## Grounding Sources\n")

    # Grounding chunks
    if chunks := metadata.grounding_chunks:
        parts.append("### Grounding Chunks\n")
        for i, chunk in enumerate(chunks, 1):
            if not (ctx := chunk.web or chunk.retrieved_context or chunk.maps):
                continue
            uri = (
                (ctx.uri or "")
                .replace("gs://", "https://storage.googleapis.com/", 1)
                .replace(" ", "%20")
            )
            parts.append(f"{i}. [{ctx.title or 'Source'}]({uri})\n")
            if getattr(ctx, "place_id", None):
                parts.append(f"    - Place ID: `{ctx.place_id}`\n\n")
            if getattr(ctx, "text", None):
                parts.append(f"{ctx.text}\n\n")

    # Queries
    if metadata.web_search_queries:
        parts.append(f"\n**Web Search Queries:** {metadata.web_search_queries}\n")
        if metadata.search_entry_point:
            parts.append(
                f"\n**Search Entry Point:**\n{metadata.search_entry_point.rendered_content}\n"
            )
    elif metadata.retrieval_queries:
        parts.append(f"\n**Retrieval Queries:** {metadata.retrieval_queries}\n")

    display(Markdown("".join(parts)))


print_grounding_data(response)
````
